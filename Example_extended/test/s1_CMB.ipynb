```julia
using PyCall

# Pythonのmatplotlib.pyplotモジュールをインポート
plt = pyimport("matplotlib.pyplot")

# plot
fig = plt.figure(figsize=(7, 6))

plt.loglog(l, Clean_map_dl[2,:], lw=2.2, label="Clean Map EE")
plt.loglog(l, Input_map_dl[2,:], lw=2.2, linestyle="dashed", label="Input Map EE")

plt.loglog(l, Clean_map_dl[3,:], lw=2.2, label="Clean Map BB")
plt.loglog(l, Input_map_dl[3,:], lw=2.2, linestyle="dashed", label="Input Map BB")

plt.ylabel(L"$\ell(\ell+1)C_\ell^{BB}/2\pi\,[\mu {\rm K}^2]$", fontsize=16)
plt.xlabel(L"$\ell$", fontsize=16)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.title("Power spectrum r=$(r)", fontsize=16)

plt.grid()
plt.legend()

# プロットを表示
plt.show()
```
