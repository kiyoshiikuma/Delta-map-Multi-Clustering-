### 領域分け index

```python
import numpy as np

# 条件を設定します
condition_1 = (beta_s.pl_index > -3.2) & (beta_s.pl_index < -3.1)
condition_2 = (beta_s.pl_index > -3.1) & (beta_s.pl_index < -3.05)
condition_3 = (beta_s.pl_index > -3.05) & (beta_s.pl_index < -3.0)
condition_4 = (beta_s.pl_index > -3.0) & (beta_s.pl_index < -2.95)
condition_5 = (beta_s.pl_index > -2.95) & (beta_s.pl_index < -2.9)
condition_6 = (beta_s.pl_index > -2.9) & (beta_s.pl_index < -2.8)

# 条件を満たす要素のインデックスを取得します
indices_1 = np.where(condition_1)[0]
indices_2 = np.where(condition_2)[0]
indices_3 = np.where(condition_3)[0]
indices_4 = np.where(condition_4)[0]
indices_5 = np.where(condition_5)[0]
indices_6 = np.where(condition_6)[0]

# インデックスを出力します
print("条件1を満たすインデックス:", indices_1)
print("条件2を満たすインデックス:", indices_2)
print("条件3を満たすインデックス:", indices_3)
print("条件4を満たすインデックス:", indices_4)
print("条件3を満たすインデックス:", indices_5)
print("条件4を満たすインデックス:", indices_6)
```

```python
array = np.zeros(len(beta_s.pl_index))   # 長さ8の配列を作成

# 指定されたインデックスに値を代入
for index in indices_1:
    array[index] = 1

for index in indices_2:
    array[index] = 2

for index in indices_3:
    array[index] = 3

for index in indices_4:
    array[index] = 4

for index in indices_5:
    array[index] = 5

for index in indices_6:
    array[index] = 6
```

```python
hp.mollview(array, title="beta_s_region", cmap='jet')
```
