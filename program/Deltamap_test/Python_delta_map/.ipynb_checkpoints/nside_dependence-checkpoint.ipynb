```python
import fg_make_file as fg
import cmb_make_file as cmb
import Delta_map as dm
import numpy as np
import healpy as hp
import warnings
warnings.filterwarnings("ignore")
```

```python
random_seed_cmb = 5123
seed_syn = 12345
nside = 2**2

r = 0.01

# rturn CMB (r = i)
# name : cmb_r_i

CMB_map = cmb.cmb_make_file(nside, r, random_seed_cmb, seed_syn)
```

### data m　作成

```python
# which_model = "s1", "d1", or "s1 and d1"
f_elements = [40, 100, 140]

data_m = dm.make_data_m(CMB_map, f_elements, "s1", nside)
```

### D matrix　計算

```python
f_elements = [40, 100, 140]

beta_elements = [-2.9, -3, -3.1, -3.01]

freq_bs, freq_bd, beta_d, T_d = 23*10**9, 353*10**9, 1.5, 20.1

D_array = []

for ii, beta_s in enumerate(beta_elements):

    D_array.append(dm.calc_D_matrix_param(f_elements, "s1", nside, freq_bs, beta_s, freq_bd, beta_d, T_d))
```

### クリーンマップ計算

```python
CMB_map_Q_array = []
CMB_map_U_array = []

#return m_[Q, U]
for ii, beta_s in enumerate(beta_elements):

    CMB_map_Q_array.append((dm.calc_clean_map(data_m, D_array[ii], nside)[0]))
    CMB_map_U_array.append((dm.calc_clean_map(data_m, D_array[ii], nside)[1]))
```

### パワースペクトル計算

```python
Clean_map_cl = []
Clean_map_dl = []
Input_map_cl = []
Input_map_dl = []

for ii, beta_s in enumerate(beta_elements):

    Clean_map_cl.append((dm.calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[0])
    Clean_map_dl.append((dm.calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[1])
    Input_map_cl.append((dm.calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[2])
    Input_map_dl.append((dm.calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[3])

    ADDD, BDDD, CDDD, DDD, l = dm.calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside)
```

### 差分パワースペクトル計算

```python
res_map_cl = []
res_map_dl = []

for ii, beta_s in enumerate(beta_elements):

    res_map_cl.append((dm.diff_calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[0])
    res_map_dl.append((dm.diff_calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[1])
    
    ADDD, BDDD, l = dm.diff_calc_spectrum(CMB_map[1], CMB_map[2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside)
```

```python
import matplotlib.pyplot as plt

#plot#
fig = plt.figure(figsize=(9, 6))

beta_elements = [-2.9, -3, -3.1, -3.01]

for ii, beta_s in enumerate(beta_elements):

    plt.loglog(l, Clean_map_dl[ii][1] , lw = 2.2, label="Clean Map EE")
    plt.loglog(l, Input_map_dl[ii][1], lw=2.2, linestyle='dashed', label="Input Map EE")

    plt.loglog(l, Clean_map_dl[ii][2] , lw = 2.2, label="Clean Map EE")
    plt.loglog(l, Input_map_dl[ii][2], lw=2.2, linestyle='dashed', label="Input Map EE")

"""  
plt.loglog(l, Clean_map_dl[7][1] , lw = 2.2, label="Clean Map EE")
plt.loglog(l, Input_map_dl[7][1], lw=2.2, linestyle='dashed', label="Input Map EE")

plt.loglog(l, Clean_map_dl[7][2] , lw = 2.2, label="Clean Map BB")
plt.loglog(l, Input_map_dl[7][2], lw = 2.2, linestyle='dashed', label="Input Map BB")
"""

plt.ylabel(r'$C_\ell^{BB}/ (\mu{\rm K}^2)$', fontsize=30)
plt.xlabel(r'$\ell$', fontsize=30)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.title("Power spectrum", fontsize=30)

plt.grid()
plt.legend()

#plt.savefig('Power_spectrum.png', bbox_inches='tight')
plt.show()
```
