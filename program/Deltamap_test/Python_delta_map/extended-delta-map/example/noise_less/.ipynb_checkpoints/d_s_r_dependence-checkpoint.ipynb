```python
import fg_make_file as fg
import cmb_make_file as cmb
import Delta_map as dm
import numpy as np
import healpy as hp
import warnings
warnings.filterwarnings("ignore")
```

```python
random_seed_cmb = 5123
seed_syn = 12345
nside = 2**2

r_elements = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001, 0.]
r_elements_name = ["1", "01", "001", "0001", "00001", "000001", "0000001", "0"]

# rturn CMB (r = i)
# name : cmb_r_i

CMB_array = []

for j, i in enumerate(r_elements):
    variable_name = f'cmb_r_{r_elements_name[j].replace(".", "_")}'
    locals()[variable_name] = cmb.cmb_make_file(nside, i, random_seed_cmb, seed_syn)

    CMB_array.append(locals()[variable_name])
```

### data m　作成

```python
# which_model = "s1", "d1", or "s1 and d1"
f_elements = [40, 50, 100, 140, 280, 402]

data_m_array = []

#return m_[Q, U]
for ii, r_in in enumerate(r_elements):

    data_m_array.append(dm.make_data_m(CMB_array[ii], f_elements, "d1 and s1", nside))
```

### D matrix　計算

```python
D = dm.calc_D_matrix(f_elements, "d1 and s1", nside)
```

### クリーンマップ計算

```python
CMB_map_Q_array = []
CMB_map_U_array = []

#return m_[Q, U]
for ii, r_in in enumerate(r_elements):

    CMB_map_Q_array.append((dm.calc_clean_map(data_m_array[ii], D, nside)[0]))
    CMB_map_U_array.append((dm.calc_clean_map(data_m_array[ii], D, nside)[1]))
```

### パワースペクトル計算

```python
Clean_map_cl = []
Clean_map_dl = []
Input_map_cl = []
Input_map_dl = []

for ii, r_in in enumerate(r_elements):

    Clean_map_cl.append((dm.calc_spectrum(CMB_array[ii][1], CMB_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[0])
    Clean_map_dl.append((dm.calc_spectrum(CMB_array[ii][1], CMB_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[1])
    Input_map_cl.append((dm.calc_spectrum(CMB_array[ii][1], CMB_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[2])
    Input_map_dl.append((dm.calc_spectrum(CMB_array[ii][1], CMB_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[3])

    ADDD, BDDD, CDDD, DDD, l = dm.calc_spectrum(cmb_r_0[1], cmb_r_0[2], CMB_map_Q_array[0], CMB_map_U_array[0], nside)
```

### 差分パワースペクトル計算

```python
res_map_cl = []
res_map_dl = []

for ii, r_in in enumerate(r_elements):

    res_map_cl.append((dm.diff_calc_spectrum(CMB_array[ii][1], CMB_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[0])
    res_map_dl.append((dm.diff_calc_spectrum(CMB_array[ii][1], CMB_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[1])
    
    ADDD, BDDD, l = dm.diff_calc_spectrum(cmb_r_0[1], cmb_r_0[2], CMB_map_Q_array[0], CMB_map_U_array[0], nside)
```
