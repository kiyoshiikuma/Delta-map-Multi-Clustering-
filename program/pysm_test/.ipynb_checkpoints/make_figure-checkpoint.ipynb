```python
import pysm3
import pysm3.units as u
import healpy as hp
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sky = pysm3.Sky(nside=2**8, preset_strings=["d1", "s1"])
```

```python
sky.components
```

```python
map_100GHz = sky.get_emission(100 * u.GHz)
```

```python
map_100GHz[0, :3]
```

```python
map_100GHz = map_100GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(100*u.GHz))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```
