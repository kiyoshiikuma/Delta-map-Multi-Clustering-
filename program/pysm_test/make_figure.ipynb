```python
import pysm3
import pysm3.units as u
import healpy as hp
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dust = pysm3.Sky(nside=2**8, preset_strings=["d0"])
synch = pysm3.Sky(nside=2**8, preset_strings=["s1"])
```

```python
dust_60GHz = dust.get_emission(40 * u.GHz)

dust_100GHz = dust.get_emission(100 * u.GHz)

dust_280GHz = dust.get_emission(140 * u.GHz)

synch_60GHz = synch.get_emission(40 * u.GHz)

synch_100GHz = synch.get_emission(100 * u.GHz)

synch_280GHz = synch.get_emission(140 * u.GHz)
```

```python
dust_60GHz = dust_60GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(60*u.GHz))

dust_1000GHz = dust_100GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(100*u.GHz))

dust_280GHz = dust_280GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(280*u.GHz))

synch_60GHz = synch_60GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(60*u.GHz))

synch_1000GHz = synch_100GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(100*u.GHz))

synch_280GHz = synch_280GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(280*u.GHz))
```

```python
Pd_60 = (dust_60GHz[1]**2 + dust_60GHz[2]**2)**(1/2)
Pd_100 = (dust_100GHz[1]**2 + dust_100GHz[2]**2)**(1/2)
Pd_280 = (dust_280GHz[1]**2 + dust_280GHz[2]**2)**(1/2) 

Ps_60 = (synch_60GHz[1]**2 + synch_60GHz[2]**2)**(1/2)
Ps_100 = (synch_100GHz[1]**2 + synch_100GHz[2]**2)**(1/2)
Ps_280 = (synch_280GHz[1]**2 + synch_280GHz[2]**2)**(1/2) 
```
