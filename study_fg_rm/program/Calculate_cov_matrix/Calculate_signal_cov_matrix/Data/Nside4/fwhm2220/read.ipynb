```python
import numpy as np

def read_cov_matrix(filename, npix):
    # バイナリファイルからデータを読み取る
    with open(filename, 'rb') as file:
        covmat_signal = np.fromfile(file, dtype=np.float64)
    return covmat_signal

filename = 'tensor_Nside4.dat'

# 共分散行列を読み取る
covmat_signal = read_cov_matrix(filename, npix)
```
