```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, platform, os
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import math
import camb
from camb import model, initialpower
print('Using CAMB %s installed at %s'%(camb.__version__,os.path.dirname(camb.__file__)))
# make sure the version and path is what you expect
```

```python
#Set up a new set of parameters for CAMB
pars = camb.read_ini('params_camb_for_PTEP.ini')
#Set WantTensors to True
pars.WantTensors = True
#This function sets up CosmoMC-like settings, with one massive neutrino and helium set using BBN consistency
pars.set_cosmology(H0=67.5, ombh2=0.022, omch2=0.122, mnu=0.06, omk=0, tau=0.06)
pars.InitPower.set_params(As=2e-9, ns=0.965, r=1)
pars.set_for_lmax(2500, lens_potential_accuracy=0);
```

```python
#r=1 cl_BB
pars.set_for_lmax(1536, lens_potential_accuracy=1)
lmax=1550

#Set WantTensors to True
pars.WantTensors = True
results = camb.get_transfer_functions(pars)
rs = 1 # r=1
inflation_params = initialpower.InitialPowerLaw()
inflation_params.set_params(ns=0.96, r=rs)
results.power_spectra_from_transfer(inflation_params) 
cl = results.get_total_cls(lmax, CMB_unit='muK')
l = np.arange(lmax+1)

#cl_lens
results = camb.get_results(pars)
powers = results.get_cmb_power_spectra(pars, CMB_unit='muK')
lensed_BB = powers['tensor'][:, 2]


#cl (TT, EE, BB, TE, EB, TB)
cl_tens = cl[:, 2]/(l*(l+1))*2*math.pi  # cl_tens

cl_lens=lensed_BB/l/(l+1)*2*math.pi   # cl_lens

#cl_hat
cl_hat = cl_lens[:lmax-14]+cl_tens[:lmax-14]*10**-1
cl_hat_s = cl_lens[:lmax-14]+cl_BB

#plot
plt.loglog(l[:1536], cl_tens[:lmax-14]+cl_lens[:lmax-14], label=r'$C_\ell$') 
plt.loglog(l[:1536], cl_hat, label=r'$\hat{C}_\ell=C_\ell^{lens}+C_\ell^{tens}\times10^{-4}$') 
plt.loglog(l[:1536], cl_hat_s,'.',label=r'$\hat{C}_\ell=C_\ell^{lens}+C_\ell^{BB}$') 
plt.loglog(l[:1536], cl_BB, label=r'$C_\ell^{BB}$') 
plt.loglog(l[:1536], cl_lens[:lmax-14], label=r'$C_\ell^{lens}$') 
plt.loglog(l[:1536], cl_tens[:lmax-14], label=r'$C_\ell^{BB}$') 


plt.ylabel(r'$C_\ell^{BB}/ (\mu{\rm K}^2)$')
plt.xlabel(r'$\ell$')
plt.legend()
```

```python
# likelihood

r = np.linspace(10**-10, 10**-9, 1000)

l_min = 2
l_max = 191

ll = np.arange(l_min,l_max+1)

L_r = np.zeros(len(r))

L_r_s = np.zeros(len(r))

#cl_hat
#cl_hat = cl_lens[:lmax-14]+cl_tens[:lmax-14]*10**-4
#cl_hat_s = cl_lens[:lmax-14]+cl_BB

for i, dr in enumerate(r):
    
    #cl
    cl = dr*cl_tens[:lmax-14]+cl_lens[:lmax-14]
    
    ln_P=np.zeros(len(ll))

    ln_P_s=np.zeros(len(ll))
    
    for  ii, l in enumerate(ll):
        
        #P_l(r)
        ln_P[ii] = -((2*l+1)/2)*(cl_hat[l]/cl[l]+np.log(cl[l])-((2*l-1)/(2*l+1))*np.log(cl_hat[l]))
        
        ln_P_s[ii] = -((2*l+1)/2)*(cl_hat_s[l]/cl[l]+np.log(cl[l])-((2*l-1)/(2*l+1))*np.log(cl_hat_s[l]))
        
        
    #L(r)
    L_r[i] = sum(ln_P)
    
    L_r_s[i] = sum(ln_P_s)

L_r = np.exp(L_r-L_r.max())

L_r_s = np.exp(L_r_s-L_r_s.max())

r_max = r[np.argmax(L_r)]

r_max_s = r[np.argmax(L_r_s)]

print(r_max)
print(r_max_s)

#plot
plt.semilogx(r, L_r)
plt.semilogx(r, L_r_s)
plt.ylabel("L(r)")
plt.xlabel("r")
plt.legend()
```
