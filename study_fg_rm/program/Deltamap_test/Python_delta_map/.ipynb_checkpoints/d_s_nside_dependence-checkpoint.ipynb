```python
import fg_make_file as fg
import cmb_make_file as cmb
import Delta_map as dm
import numpy as np
import healpy as hp
import warnings
warnings.filterwarnings("ignore")
```

```python
random_seed_cmb = 5123
seed_syn = 12345
nside_elements = [4, 8, 16]

r = 0.01

# rturn CMB (r = i)
# name : cmb_r_i

CMB_map_array =[]

for ii, nside in enumerate(nside_elements):
    
    CMB_map = cmb.cmb_make_file(nside, r, random_seed_cmb, seed_syn)

    CMB_map_array.append(CMB_map)
```

### data m　作成

```python
# which_model = "s1", "d1", or "s1 and d1"
f_elements = [40, 100, 140]

data_m_array = []

for ii, nside in enumerate(nside_elements):
    
    data_m = dm.make_data_m(CMB_map_array[ii], f_elements, "s1", nside)

    data_m_array.append(data_m)
```

### D matrix　計算

```python
f_elements = [40, 100, 140]

beta_s = -3.

freq_bs, freq_bd, beta_d, T_d = 23*10**9, 353*10**9, 1.5, 20.1

D_array = []

for ii, nside in enumerate(nside_elements):

    D_array.append(dm.calc_D_matrix_param(f_elements, "s1", nside, freq_bs, beta_s, freq_bd, beta_d, T_d))
```

### クリーンマップ計算

```python
CMB_map_Q_array = []
CMB_map_U_array = []

#return m_[Q, U]
for ii, nside in enumerate(nside_elements):

    CMB_map_Q_array.append((dm.calc_clean_map(data_m_array[ii], D_array[ii], nside)[0]))
    CMB_map_U_array.append((dm.calc_clean_map(data_m_array[ii], D_array[ii], nside)[1]))
```

### パワースペクトル計算

```python
Clean_map_cl = []
Clean_map_dl = []
Input_map_cl = []
Input_map_dl = []
l =[]

for ii, nside in enumerate(nside_elements):

    Clean_map_cl.append((dm.calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[0])
    Clean_map_dl.append((dm.calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[1])
    Input_map_cl.append((dm.calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[2])
    Input_map_dl.append((dm.calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[3])
    l.append((dm.calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[4])
```

### 差分パワースペクトル計算

```python
res_map_cl = []
res_map_dl = []

for ii, nside in enumerate(nside_elements):

    res_map_cl.append((dm.diff_calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[0])
    res_map_dl.append((dm.diff_calc_spectrum(CMB_map_array[ii][1], CMB_map_array[ii][2], CMB_map_Q_array[ii], CMB_map_U_array[ii], nside))[1])
```

```python
import matplotlib.pyplot as plt

#plot#
fig = plt.figure(figsize=(9, 6))

for ii, nside in enumerate(nside_elements):

    plt.loglog(l[ii], Clean_map_dl[ii][1] , lw = 2.2, label=f'{"Clean EE nside="}{str(nside)}')
    plt.loglog(l[ii], Input_map_dl[ii][1], lw=2.2, linestyle='dashed', label=f'{"Input EE nside="}{str(nside)}')

    plt.loglog(l[ii], Clean_map_dl[ii][2] , lw = 2.2, label=f'{"Clean BB nside="}{str(nside)}')
    plt.loglog(l[ii], Input_map_dl[ii][2], lw=2.2, linestyle='dashed', label=f'{"Input BB nside="}{str(nside)}')

"""  
plt.loglog(l, Clean_map_dl[7][1] , lw = 2.2, label="Clean Map EE")
plt.loglog(l, Input_map_dl[7][1], lw=2.2, linestyle='dashed', label="Input Map EE")

plt.loglog(l, Clean_map_dl[7][2] , lw = 2.2, label="Clean Map BB")
plt.loglog(l, Input_map_dl[7][2], lw = 2.2, linestyle='dashed', label="Input Map BB")
"""

plt.ylabel(r'$Δ \ell(\ell+1)C_\ell^{BB}/2\pi\,[\mu {\rm K}^2]$', fontsize=30)
plt.xlabel(r'$\ell$', fontsize=30)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.title("Power spectrum", fontsize=30)

plt.grid()
plt.legend(fontsize=13)

#plt.savefig('Power_spectrum.png', bbox_inches='tight')
plt.show()
```
