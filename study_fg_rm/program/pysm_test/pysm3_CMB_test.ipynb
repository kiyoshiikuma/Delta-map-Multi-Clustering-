```python
import pysm3
import pysm3.units as u
import healpy as hp
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sky = pysm3.Sky(nside=2**8, preset_strings=["c1"])
```

```python
sky.components
```

```python
map_100GHz = sky.get_emission(100 * u.GHz)
map_140GHz = sky.get_emission(400 * u.GHz)
```

```python
map_100GHz[0, :3]
map_140GHz[0, :3]
```

```python
map_100GHz = map_100GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(100*u.GHz))
map_140GHz = map_140GHz.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(140*u.GHz))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#print(map_100GHz[0])
#print(map_140GHz[0])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 6))

plt.axes(ax1)
hp.mollview(map_100GHz[0], title=" map", hold=True,unit = map_100GHz.unit)
plt.axes(ax2)
hp.mollview(map_100GHz[1], title=" map", hold=True,unit = map_100GHz.unit)
plt.axes(ax3)
hp.mollview(map_100GHz[2], title=" map", hold=True,unit = map_100GHz.unit)

#hp.mollview(map_140GHz[0], title=" map", unit = map_140GHz.unit)
#hp.mollview(map_100GHz[1] - map_140GHz[1], title=" map", unit = map_140GHz.unit)
```

```python
smoothed_data = hp.smoothing(map_100GHz[0], fwhm=1)
```

```python
hp.mollview(map_100GHz[0], title=" map", unit = map_100GHz.unit)
hp.mollview(smoothed_data, title=" map", unit = map_100GHz.unit)
```

```python
hp.mollview(map_100GHz[0], title=" map", min=-3*10**6, max=3*10**6, unit = map_100GHz.unit)
hp.mollview(smoothed_data, title=" map", min=-3*10**6, max=3*10**6, unit = map_100GHz.unit)
```
